# Problem Statement:

Objective:
    
The objective of this project is to analyze the Google Play Store dataset to uncover insights related to app popularity, 
user ratings, pricing strategies, and download patterns. The goal is to identify key factors that influence app success,
such as category trends, app size, content rating, and user feedback. Based on these insights, actionable recommendations 
will be provided for app developers and marketers to improve app visibility, engagement, and profitability on the platform.


Task:
    
1.Data Cleaning:

Handle missing values, remove duplicates, and standardize data formats (e.g., ratings, installs, and prices).

2.Exploratory Data Analysis (EDA):

Analyze app distribution across categories, rating patterns, and the relationship between key variables like price,
installs.

3.Trend Analysis:

Identify top categories by downloads, compare free vs. paid apps, and explore the impact of app updates on performance.

4.Correlation Analysis:

Investigate relationships between key factors (e.g., app size, pricing, ratings, reviews) to understand what drives
app success.

5.Actionable Insights & Recommendations:

Provide recommendations to app developers and marketers based on findings, such as ideal pricing, category selection,
and update strategies.



Dataset: Google Play Store
    
Dataset Discription:
    
App Name         :   The name of the application listed on the Google Play Store.

App Id           :   A unique identifier assigned to each app in the Google Play Store.

Category         :   The category under which the app is listed (e.g., Education, Tools, Business, etc.).

Rating           :   Average user rating of the app (range: 1 to 5).

Rating Count     :   The number of user ratings provided for the app.

Installs         :   Total number of times the app has been installed (in ranges like 1K+, 10K+, etc.).

Minimum Installs :   The minimum number of installations for the app.

Maximum Installs :   The maximum number of installations for the app.

Free             :   Indicates whether the app is free (True/False).

Price            :   The price of the app (in USD), if it is a paid app.

Currency         :   The currency in which the price is displayed.

Size             :   The size of the app (in MB).

Minimum Android  :   The minimum Android version required to run the app.

Developer Id     :   A unique identifier for the developer of the app.

Developer Website:   The official website of the app’s developer.

Developer Email  :   The contact email address of the developer.

Released         :   The date when the app was originally released on the Play Store.

Last Updated     :   The date when the app was last updated on the Play Store.

Content Rating   :   The age group targeted by the app (e.g., Everyone, Teen, Mature 17+).

Privacy Policy   :   The URL linking to the app’s privacy policy.

Ad Supported     :   Indicates if the app is supported by ads (True/False).

In App Purchases :   Indicates if the app offers in-app purchases (True/False).

Editors Choice   :   Indicates whether the app has been selected as an Editors’ Choice app (True/False).

Scraped Time     :   The timestamp when the app data was last scraped from the Google Play Store.

```python
# Perform Data Cleaning
```

```python
# Handle missing values
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df1=df.dropna()
```

```python
df1.isnull().sum()
```

```python
# Remove duplicates
```

```python
df=df1.drop_duplicates(subset="App Id")
```

```python
df.head()
```

```python
df.index
```

```python
# standardize data formats
```

```python
df["Installs"] = df["Installs"].str.replace(",", "")  
df["Installs"] = df["Installs"].str.replace("+", "")  
df["Installs"] = pd.to_numeric(df["Installs"])  
```

```python
df["Released"]=pd.to_datetime(df["Released"])
df["Last Updated"]=pd.to_datetime(df["Last Updated"])
df["Scraped Time"]=pd.to_datetime(df["Scraped Time"])
```

```python
def convert_size(size):
    if isinstance(size, str):
        size = size.replace(',', '')
        if 'M' in size:
            return float(size.replace('M', '')) 
        elif 'k' in size:
            return float(size.replace('k', '')) / 1000 
    return np.nan 

df['Size'] = df['Size'].apply(convert_size)
```

```python
df.info()
```

```python
# Perform Exploratory Data Analysis
```

```python
df.head()
```

```python
# App Distribution Across Categories
```

```python
App_count_category=df.groupby("Category")["App Id"].count().reset_index()
App_count_cat_sort=App_count_category.sort_values(by="App Id",ascending=False)
App_count_cat_sort
```

```python
plt.figure(figsize=(14,5))
sns.barplot(x="Category",y="App Id",data=App_count_cat_sort,palette='viridis')
plt.title("Count of App per Category")
plt.xticks(rotation=90)
plt.show()
```

```python
# Rating Patterns Across Categories
```

```python
Avg_Rating_cat=df.groupby('Category')['Rating',"Installs"].mean().reset_index()
Avg_Rating_cat_sort=Avg_Rating_cat.sort_values(by=["Rating","Installs"],ascending=False)
Avg_Rating_cat_sort
```

```python
plt.figure(figsize=(14,5))
# sns.set(style="whitegrid")
sns.barplot(x="Category",y="Rating",data=Avg_Rating_cat_sort,palette='viridis')
plt.title("Average Rating by Category")
plt.xticks(rotation=90)
plt.show()
```

```python
# Top Free vs. Paid Apps
```

```python
# Top installed Free Apps
```

```python
df_free=df[df["Price"]==0]
```

```python
df_free_sort=df_free.sort_values(by="Installs",ascending=False)
df_free_sort[["App Name","Rating",'Installs']].head(10)
```

```python
# Compare free app rating and Installs 
```

```python
avg_free_rating=df_free["Rating"].mean()
avg_free_Installs=df_free["Installs"].mean()
print("avg_free_rating:",avg_free_rating)
print("avg_free_Installs:",avg_free_Installs)
```

```python
# Top Paid Apps
```

```python
df_paid=df[df["Price"]>0]
```

```python
df_paid_sort=df_paid.sort_values(by="Price",ascending=False)
df_paid_sort[["App Name","Price","Installs","Rating"]].head(10)
```

```python
#  Compare Paid app rating and Installs 
```

```python
avg_paid_rating = df_paid['Rating'].mean()
avg_paid_Installs = df_paid['Installs'].mean()
print("avg_paid_rating:",avg_paid_rating)
print("avg_paid_Installs",avg_paid_Installs)
```

```python
# Average rating comparision free app and paid app
```

```python
Categories=["Free Apps","Paid Apps"]
Average_ratings=[avg_free_rating,avg_paid_rating]

plt.figure(figsize=(8, 6))
ax=sns.barplot(x=Categories, y=Average_ratings)
ax.bar_label(ax.containers[0])
plt.title('Average Ratings: Free vs. Paid Apps')
plt.ylabel('Average Rating')
plt.show()
```

```python
Categories=["Free Apps","Paid Apps"]
Average_Installs=[avg_free_Installs,avg_paid_Installs]

plt.figure(figsize=(8, 6))
ax=sns.barplot(x=Categories,y=Average_Installs)
ax.bar_label(ax.containers[0])
plt.title('Average Installs: Free vs. Paid Apps')
plt.ylabel('Average Installs')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Price', y='Rating', data=df_paid)

plt.title('Price vs. Ratings for Paid Apps')
plt.xlabel('Price')
plt.ylabel('Rating')

plt.show()
```

```python
# Time-Based Analysis
```

```python
plt.figure(figsize=(13, 6))
sns.lineplot(x='Released', y='Installs', data=df)
plt.yscale('log')
plt.title('Released Date vs. Installs')
plt.xlabel('Released Date')
plt.ylabel('Installs')

plt.show()
```

```python
df['Released Year'] = df["Released"].dt.year
```

```python
performance_by_year_r = df.groupby('Released Year').agg({
    'Rating': 'mean',
}).reset_index()

print(performance_by_year_r)
```

```python
df['Last Updated Year'] = df['Last Updated'].dt.year
```

```python
performance_by_year = df.groupby('Last Updated Year').agg({
    'Rating': 'mean',
}).reset_index()

print(performance_by_year)
```

```python
plt.figure(figsize=(12, 6))


plt.plot(performance_by_year_r['Released Year'], performance_by_year_r['Rating'], 
         marker='o', label='Average Rating by Released Year', color='blue')

plt.plot(performance_by_year['Last Updated Year'], performance_by_year['Rating'], 
         marker='x', label='Average Rating by Last Updated Year', color='orange')

plt.title('Average Ratings Over Time (Released vs. Last Updated Year)')
plt.xlabel('Year')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```

# Recommendations:

•Category Selection:
    
Focus on Racing, Action, and Video Players & Editors for high installs and visibility.
Boost Role Playing, Casino, and Simulation through marketing, as they have high ratings but moderate installs.
Avoid Food & Drink, Beauty, and Events due to low installs and poor ratings.

•Pricing Strategy:

Prioritize free apps with in-app purchases or ads for better market reach.
For paid apps, lower price points or offer trial versions to increase installs.
Reconsider pricing strategies for high-priced apps with low installs; reduce prices or bundle more features.

•App Update Strategies:
    
Release frequent updates to improve user satisfaction and app ratings.
Clearly communicate updates to engage existing users and attract new ones.

•Marketing Insights:
    
Promote high-rating apps in marketing efforts to increase visibility and downloads.
Use positive user reviews in categories like Role Playing and Casino for promotional content.

•User Feedback & Experience:
    
Continuously address user feedback, especially in low-performing categories like Beauty and Food & Drink.
Build a strong user community by engaging with users regularly to improve retention.

•Data-Driven Decision Making:
    
Use category trends and user sentiment data for app launches and updates.
Focus on improving app quality and marketing, as price, installs, and ratings show little correlation.
